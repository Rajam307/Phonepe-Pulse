```python
import os
import json
import pandas as pd
import psycopg2
from sqlalchemy import create_engine
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
## AGGREGATED TRANSACTION ##

path1 = "C:/Users/hp/Documents/RAJI/pulse/data/aggregated/transaction/country/india/state/"
agg_trans_list = os.listdir(path1)

Table1={"States":[], "Years":[], "Quarters":[], "Transaction_type":[], "Transaction_count":[], "Transaction_amount":[]}

for state in agg_trans_list:
    cur_states = path1+state+"/"  # print(cur_states)
    agg_year_list = os.listdir(cur_states)   #print(agg_year_list)

    for year in agg_year_list:
        cur_years =cur_states+year+"/" #print(cur_years)
        agg_file_list = os.listdir(cur_years)  # print(agg_file_list)

        for file in  agg_file_list:
            quar_files = cur_years+file
            data=open(quar_files,"r")

            AT=json.load(data)
            
            for i in AT["data"]["transactionData"]:
                    name=i["name"]   #print(name)
                    count=i["paymentInstruments"][0]["count"]
                    amount=i["paymentInstruments"][0]["amount"]
                    Table1["States"].append(state)
                    Table1["Years"].append(year)
                    Table1["Quarters"].append(int(file.strip(".json")))
                    Table1["Transaction_type"].append(name)
                    Table1["Transaction_count"].append(count)
                    Table1["Transaction_amount"].append(amount)
            


```

```python
AT_table1=pd.DataFrame(Table1)
AT_table1["States"]=AT_table1["States"].str.replace('andaman-&-nicobar-islands','Andaman & Nicobar')
AT_table1["States"]=AT_table1["States"].str.replace("-"," ")
AT_table1["States"]=AT_table1["States"].str.title()
AT_table1["States"]=AT_table1["States"].str.replace('Dadra & Nagar Haveli & Daman & Diu','Dadra and Nagar Haveli and Daman and Diu',)
```
